```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from iminuit import Minuit
```

```python
sns.set_style("darkgrid", {"grid.color": "0", "grid.linestyle": ":"})
plt.rcParams['image.cmap'] = 'hsv'
plt.rcParams['font.size'] = 15
```

```python
cores = np.array([1,2,4,8,16,32,64])
time = np.array([0.8386,0.9245,1.023,1.288,1.586,2.676,4.8])
```

```python
time[0]/time
```

```python
def gus(x,p):
    return 1-p+p*x
```

```python
def chi2(func, x, y, stderr, *args):
    return np.sum((y-func(x,*args))**2/stderr**2)
```

```python
chi = lambda p: chi2(gus, cores, time/time[0], 1, p)
ll = Minuit(chi,p=0.5)
ll.limits = [(0,1)]
ll.migrad()
```

```python
times = np.array([0.8272,0.952,1.169,1.609,2.331,3.795,6.779])
```

```python
chi = lambda p: chi2(gus, cores, times/times[0], 1, p)
ll = Minuit(chi,p=0.5)
ll.limits = [(0,1)]
ll.migrad()
```

```python
plt.plot(cores,time/time[0],'.')
plt.plot(cores,gus(cores,0.074),'-',label="With FFT-opt: P=0.074")
plt.plot(cores,times/times[0],'.')
plt.plot(cores,gus(cores,0.115),'-',label="Without FFT-opt: P=0.115")
plt.xlabel("Number of cores")
plt.ylabel("Scaled Speedup(N)")
plt.legend()
plt.xlim(0,20)
#plt.savefig('weakscaling.eps')
```
