```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from iminuit import Minuit
import seaborn as sns
```

```python
sns.set_style("darkgrid", {"grid.color": "0", "grid.linestyle": ":"})
plt.rcParams['image.cmap'] = 'hsv'
plt.rcParams['font.size'] = 15
```

```python
def ahmdalslaw(x,p):
    return 1/(1-p+p/x)
```

```python
def chi2(func, x, y, stderr, *args):
    return np.sum((y-func(x,*args))**2/stderr**2)
```

```python
#small1
#cores_small1 = np.array([1, 2, 4, 8, 16, 24, 32, 40, 48, 64])
#times_small1 = np.array([1.42976, 0.901623, 0.574469, 0.456935, 0.430265, 0.412449, 0.381068, 0.365572, 0.340546, 0.37666]) #sec
#times_small1 = times_small[0] / times_small1

#small2
cores_small2 = np.array([1, 2, 4, 8, 16, 24, 32, 40, 48, 64])
times_small2 = np.array([13.5003, 6.44922, 4.96258, 4.11658, 3.40945, 3.38683, 2.96005, 3.04346, 2.90602, 2.74532])
times_small2 = times_small2[0] / times_small2
cores_small3 = np.array([1, 2, 4, 8, 16, 32, 40, 48, 64])
times_small3 = np.array([13.5003, 6.44922, 4.96258, 4.11658, 3.40945, 2.96005, 3.04346, 2.90602, 2.74532])
times_small3 = times_small3[0] / times_small3

#medium
cores_medium = np.array([1, 2, 4, 8, 16, 24, 32, 40, 48, 64])[:-2]
times_medium = np.array([180.264, 119.209, 69.9604, 47.2896, 37.8779, 34.9556, 33.3372, 32.1254, 37.858, 37.9116])[:-2]
times_medium = times_medium[0] / times_medium

#large
cores_large = np.array([1, 2, 4, 8, 16, 24, 32, 40, 48, 64])[:-3]
times_large = np.array([70.7019, 43.1892, 38.4618, 23.2511, 20.2951, 19.8368, 18.8477, 22.6592, 24.7627, 34.5212])[:-3]
times_large = times_large[0] / times_large
```

```python
time_2 = np.array([1.25472, 3.1875, 4.21975, 5.66163])
gangs2 = np.array([1, 4, 16, 64])
time_2 = (time_2[0]*gangs2)/time_2
```

```python
chi = lambda p: chi2(ahmdalslaw, cores_small2, times_small2, 1, p)
ll = Minuit(chi,p=0.9)
ll.migrad()
```

```python
chi = lambda p: chi2(ahmdalslaw, cores_small3, times_small3, 1, p)
ll = Minuit(chi,p=0.9)
ll.migrad()
```

```python
chi = lambda p: chi2(ahmdalslaw, cores_medium, times_medium, 1, p)
ll = Minuit(chi,p=0.6)
ll.migrad()
```

```python
chi = lambda p: chi2(ahmdalslaw, cores_large, times_large, 1, p)
ll = Minuit(chi,p=0.5)
ll.migrad()
```
